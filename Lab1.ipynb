```python
# Imports:
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris

# Load iris dataset so we can use it:
iris = load_iris()
```

```python
# Question 1:

# Assign feature names:
f0 = iris['feature_names'][0]
f1 = iris['feature_names'][1]
f2 = iris['feature_names'][2]
f3 = iris['feature_names'][3]

# Assign axis values:
X = iris['data']
y = iris['target']

# Create figure:
fig, ax = plt.subplots(3, 2, figsize=(15, 15))
plt.suptitle("Question 1")

# Plot graphs:
ax[0,0].scatter(X[:,0], X[:, 1], c=y, s=60)
ax[0,0].set_xlabel(f0)
ax[0,0].set_ylabel(f1)

ax[1,0].scatter(X[:,0], X[:, 2], c=y, s=60)
ax[1,0].set_xlabel(f0)
ax[1,0].set_ylabel(f2)

ax[2,0].scatter(X[:,0], X[:, 3], c=y, s=60)
ax[2,0].set_xlabel(f0)
ax[2,0].set_ylabel(f3)

ax[0,1].scatter(X[:,1], X[:, 2], c=y, s=60)
ax[0,1].set_xlabel(f1)
ax[0,1].set_ylabel(f2)

ax[1,1].scatter(X[:,1], X[:, 3], c=y, s=60)
ax[1,1].set_xlabel(f1)
ax[1,1].set_ylabel(f3)

ax[2,1].scatter(X[:,2], X[:, 3], c=y, s=60)
ax[2,1].set_xlabel(f2)
ax[2,1].set_ylabel(f3)
```

```python
# Conclusion
print('Feature 0 = ' + f0)
print('Feature 1 = ' + f1)
print('Feature 2 = ' + f2)
print('Feature 3 = ' + f3)
```

```python
# Question 2:

fig, ax = plt.subplots(3, 3, figsize=(15, 15))
plt.suptitle("iris_pairplot")

for i in range(3): # Go from 0-2
    for j in range(3): # Go from 0-2
        # Thus iterate through all subplots in ax
        
        # Plot the scatter for feature i against feature j:
        ax[i,j].scatter(X[:,j], X[:,i+1], c=y, s=60)
        
        # The colors represent different iris types
        ax[i,j].set_xticks(())
        ax[i,j].set_yticks(())
        
        if i == 2: # For the bottom row, set the x label
            ax[i,j].set_xlabel(iris['feature_names'][j])
            
        if j == 0: # For the left-most column, set the y label
            ax[i,j].set_ylabel(iris['feature_names'][i+1])
            
        if j > i: # Hide inverted repeats
            ax[i,j].set_visible(False)
```
